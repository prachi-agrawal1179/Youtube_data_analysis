```python
#import all necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'/content/top_200_youtubers.csv')
```

```python
data
```

```python
data.columns
```

```python
#checking all data types of each columns
data.dtypes
```

```python
#checking for null values in each column
data.info()
```

```python
#dropping duplicates
data = data.drop_duplicates().reset_index(drop=True)
data.info()
```

```python
#filling all missing values with mean values 

data['Avg. 1 Day'] = data['Avg. 1 Day'].fillna(data['Avg. 1 Day'].mean())
data['Avg. 3 Day'] = data['Avg. 3 Day'].fillna(data['Avg. 3 Day'].mean())
data['Avg. 7 Day'] = data['Avg. 7 Day'].fillna(data['Avg. 7 Day'].mean())
data['Avg. 14 Day'] = data['Avg. 14 Day'].fillna(data['Avg. 14 Day'].mean())
data['Avg. 30 day'] = data['Avg. 30 day'].fillna(data['Avg. 30 day'].mean())
data.describe()
```

```python
data.info()
```

```python
#filling missing values with mean value

data['Engagement Rate'] = data['Engagement Rate'].fillna(data['Engagement Rate'].mean())
data['Views Avg.'] = data['Views Avg.'].fillna(data['Views Avg.'].mean())
data['Comments Avg'] = data['Comments Avg'].fillna(data['Comments Avg'].mean())
data.info()
```

```python
#checking relation of missing values of category column with main video category

display(data[data['Category'].isna() == True])
```

```python
#checking for all unique values of main category column 

data['Main Video Category'].unique()
```

```python
data['Country'] = data['Country'].fillna('None')
data.info()
```

```python
#channels that have Music or Pop music in the Main Video Category can be classified as Music. 
#Similarly, Gaming, Strategy video game and Action-adventure game 
#can be categorized as Gaming & Apps, and Sports to Sports. In addition, 
#with a big stretch, Education can be attributed to Science & Tech

data.loc[((data['Category'].isna() == True) | (data['Category']=='None')) &
            ((data['Main Video Category']=='Music') | 
             (data['Main Video Category']=='Pop music')),'Category'] = 'Music'

data.loc[((data['Category'].isna() == True) | (data['Category']=='None')) &
           ((data['Main Video Category']=='Gaming') | 
            (data['Main Video Category']=='Strategy video game') | 
            (data['Main Video Category']=='Action-adventure game')),'Category'] = 'Gaming & Apps'

data.loc[((data['Category'].isna() == True) | (data['Category']=='None')) & 
            (data['Main Video Category']=='Sports'),'Category'] = 'Sports'

data.loc[((data['Category'].isna() == True) | (data['Category']=='None')) & 
           (data['Main Video Category']=='Education'),'Category'] = 'Science & Tech'

data.loc[((data['Category'].isna() == True) | (data['Category']=='None')) & 
           ((data['Main Video Category']=='People & Blogs') | 
            (data['Main Video Category']=='Howto & Style')),'Category'] = 'LifeStyle'


data.loc[((data['Category'].isna() == True) | (data['Category']=='None')) & 
           ((data['Main Video Category']=='Shows') | 
            (data['Main Video Category']=='TV shows') | 
            (data['Main Video Category']=='Entertainment') |
            (data['Main Video Category']=='Comedy') | 
            (data['Main Video Category']=='Film & Animation')),'Category'] = 'Entertainment'

data.loc[((data['Category'].isna() == True) | (data['Category']=='None')) & 
           ((data['Main Video Category']=='News & Politics') | 
            (data['Main Video Category']=='Nonprofits & Activism')),'Category'] = 'Other'           
```

```python
data.info()
```

```python
#checking which country have largest youtube channels

country = data.Country.value_counts().nlargest(10)
plt.figure(figsize =(10, 7), dpi=100)
country.plot.bar(color='#ff0000')
plt.title("Country", fontsize=18)
plt.show()
```

```python
#checking which channel have largest youtube followers

followers = data.groupby('Channel Name')['followers'].sum().nlargest(10)
plt.figure(figsize =(10, 7), dpi=100)
followers.plot.bar(color='#0000ff')
plt.title("Channels with highest number of followers", fontsize=18)
plt.show()
```

```python
#checking which category have largest youtube followers

followers_category = data.groupby('Category')['followers'].sum().nlargest(10)
plt.figure(figsize =(10, 7), dpi=100)
followers_category.plot.bar(color='#0000ff')
plt.title("Category with highest number of followers", fontsize=18)
plt.show()
```

```python
#checking which channel have largest no of likes

Likes = data.groupby('Channel Name')['Likes'].sum().nlargest(10)
plt.figure(figsize =(10, 7), dpi=100)
Likes.plot.bar(color='#0000ff')
plt.title("Channels with highest number of Likes", fontsize=18)
plt.show()
```

```python
#checking which category have largest no of likes

likes_category = data.groupby('Category')['Likes'].sum().nlargest(10)
plt.figure(figsize =(10, 7), dpi=100)
likes_category.plot.bar(color='#0000ff')
plt.title("Category with highest number of likes", fontsize=18)
plt.show()
```

```python
#checking which channel have largest no of engagement rate

engagement_rate = data.groupby('Channel Name')['Engagement Rate'].sum().nlargest(10)
plt.figure(figsize =(10, 7), dpi=100)
engagement_rate.plot.bar(color='#0000ff')
plt.title("Channels with highest engagement rate", fontsize=18)
plt.show()
```

```python
#checking which category have largest no of engagement rate

engagement_rate_category = data.groupby('Category')['Engagement Rate'].sum().nlargest(10)
plt.figure(figsize =(10, 7), dpi=100)
engagement_rate_category.plot.bar(color='#0000ff')
plt.title("Category with highest Engagement Rate", fontsize=18)
plt.show()
```

```python
#checking which channel have largest no of views

views = data.groupby('Channel Name')['Views'].sum().nlargest(10)
plt.figure(figsize =(10, 7), dpi=100)
views.plot.bar(color='#0000ff')
plt.title("Channels with highest views", fontsize=18)
plt.show()
```

```python
#checking which category have largest no of views

views_category = data.groupby('Category')['Views'].sum().nlargest(10)
plt.figure(figsize =(10, 7), dpi=100)
views_category.plot.bar(color='#0000ff')
plt.title("Category with highest views", fontsize=18)
plt.show()
```

```python
#checking which channel have largest no of comments

comments = data.groupby('Channel Name')['Comments Avg'].sum().nlargest(10)
plt.figure(figsize =(10, 7), dpi=100)
comments.plot.bar(color='#0000ff')
plt.title("Channels with highest comments", fontsize=18)
plt.show()
```

```python
#checking which category have largest no of comments

comments_category = data.groupby('Category')['Comments Avg'].sum().nlargest(10)
plt.figure(figsize =(10, 7), dpi=100)
comments_category.plot.bar(color='#0000ff')
plt.title("Category with highest comments", fontsize=18)
plt.show()
```

```python
#check views and likes based on genre(main topic of channel)

more_likes_genre = data.groupby('Main topic')['Likes'].sum().nlargest(5)
more_views_genre = data.groupby('Main topic')['Views'].sum().nlargest(5)

plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.subplots_adjust(wspace=0.3, hspace=0)
sns.barplot(data=data, x=more_likes_genre, y=more_likes_genre.index, color='#CABBE9')
plt.title('Genres with most likes')
plt.xlabel('')
plt.ylabel('')
plt.xticks([0.00e+09,2.00e+09,4.00e+09,6.00e+09,8.00e+09],labels = ['0M','2 Bn','4 Bn','6 Bn','8 Bn'])

plt.subplot(1,2,2)
sns.barplot(data=data, x=more_views_genre, y=more_views_genre.index, color='#CABBE9')

plt.title('Genres with most views')
plt.xlabel('')
plt.ylabel('')
plt.xticks([0.00e+11,2.00e+11,4.00e+11,6.00e+11,8.00e+11],labels = ['0 M','200 Bn','400 Bn','600 Bn','800 Bn'])
plt.show()
```
